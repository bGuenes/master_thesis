```python
from Chempy.wrapper import Chempy
from Chempy.parameter import ModelParameters

import numpy as np
import matplotlib.pyplot as plt
```

```python
a = ModelParameters()
```

```python
a.solar_abundance_name = "Asplund05_apogee_correction"
a.yield_table_name_1a = "Thielemann"
```

```python
cube, abun = Chempy(a)
```

```python
elements = abun.dtype.names

for i in elements[0:10]:
    plt.plot(abun['time'], np.exp(abun[i]), label=i)
plt.legend()
```

```python
plt.plot(abun['Fe'][1:],abun['O'][1:]-abun['Fe'][1:], label = 'O')
plt.plot(abun['Fe'][1:],abun['Mn'][1:]-abun['Fe'][1:], label = 'Mn')
plt.plot(abun['Fe'][1:],abun['N'][1:]-abun['Fe'][1:], label = 'N')
plt.xlabel('[Fe/H]')
plt.ylabel('[X/Fe]')
plt.legend()
```
